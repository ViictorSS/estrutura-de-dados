```python
%%writefile exercicio-3.c

 //isalpha (char c) retorna 1 se c é uma letra e 0 no caso contrário.
 //isdigit (char c) retorna 1 se c é um dígito e 0 no caso contrário.
 //isupper (char c) retorna 1 se c é uma letra maiúscula e 0 no caso contrário.
 //islower (char c) retorna 1 se c é uma letra minúscula e 0 no caso contrário.
 //isalnum (char c) retorna 1 se c é uma letra ou um dígito e 0 no caso contrário




#include <stdio.h>
#include <ctype.h>

int main()
{
   char c;
   c = 'I';
   printf("Retorna não 0 se for letra: %d\n", isalpha(c));

   c = '2';
   printf("Retorna não 0 se for digito: %d\n", isdigit(c));

   c='C';
   printf("Retorna não 0 se for letra maiúscula: %d\n", isupper(c));

   c='c';
   printf("Retorna não 0 se for letra minúscula: %d\n", islower(c));

   c='C';
   printf("Retorna não 0 se for alfa-numerico: %d\n", isalnum(c));

   return 0;
}
```
